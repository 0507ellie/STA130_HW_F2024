- linear regression: Yi=β0+β1xi+ϵi
    - Yi is the so-called outcome. This is the dependent variable we’re trying to predict.
    - xi is then the predictor. This is the independent variable used to predict the value of Y.
    - β0 is the intercept coefficient. The intercept represents the expected value of Y when X=0.
    - β1 is the slope coefficient. This coefficient shows how much Y is expected to change with a one-unit increase in xi.
    - ϵi is called the error term. This term represents the difference between the observed value of Y and the value predicted by the linear model.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

np.random.seed(42)  
beta_0 = 5        # Intercept
beta_1 = 2.5      # Slope
sigma = 1.5       # Standard deviation of the error term
n = 100           # Number of samples

# Generate predictor variable X as random values
X = np.random.rand(n) * 10  # X ranges from 0 to 10

# Generate the error term epsilon from a normal distribution
epsilon = np.random.normal(0, sigma, n)

# Compute the outcome variable Y based on the SLR model
Y = beta_0 + beta_1 * X + epsilon

# Plotting the result to visualize the model
plt.figure(figsize=(10, 6))
plt.scatter(X, Y, color='blue', label="Observed data (Y)")
plt.plot(X, beta_0 + beta_1 * X, color='red', label="True Regression Line (without error)")
plt.xlabel("Predictor Variable (X)")
plt.ylabel("Outcome Variable (Y)")
plt.title("Simple Linear Regression Model Simulation")
plt.legend()
plt.show()

# Displaying the generated data's mean and standard deviation
print("Mean of Y values (outcome):", np.mean(Y))
print("Standard deviation of Y values:", np.std(Y))
````

````python
import numpy as np
import pandas as pd
import plotly.express as px
import statsmodels.formula.api as smf  # for fitting linear models

# Step 1: Simulate data
np.random.seed(42)  # for reproducibility
beta_0 = 5          # intercept
beta_1 = 2.5        # slope
sigma = 1.5         # standard deviation of the error term
n = 100             # number of samples

# Generate predictor variable X and outcome variable Y
X = np.random.rand(n) * 10  # X values between 0 and 10
epsilon = np.random.normal(0, sigma, n)  # error term
Y = beta_0 + beta_1 * X + epsilon  # create outcome variable based on the linear model

# Step 2: Combine X and Y into a pandas DataFrame named df
df = pd.DataFrame({'x': X, 'Y': Y})

# Step 3: Specify and fit the model
# What are the following two steps doing?
model_data_specification = smf.ols("Y ~ x", data=df)  # Specifies an OLS model with Y as the outcome and x as the predictor
fitted_model = model_data_specification.fit()         # Fits the model to find best estimates for beta_0 and beta_1

# Step 4: Extract information from the fitted model
# fitted_model.summary() 
# Provides a comprehensive summary of the model, including coefficient estimates, their statistical significance, and overall model fit statistics

# fitted_model.summary().tables[1] explanation
# Shows a summary table specifically for the regression coefficients (beta_0 and beta_1), including their values, standard errors, and p-values

# fitted_model.params explanation
# Returns a pandas Series with estimated values of the intercept (beta_0) and slope (beta_1)

# fitted_model.params.values explanation
# Provides just the numerical values of the intercept and slope in an array format (useful for computation)

# fitted_model.rsquared explanation
# Provides the R-squared value, which measures the proportion of variance in Y explained by the model with X

# Display the model summary and extracted components for review
print(fitted_model.summary())
print(fitted_model.summary().tables[1])
print("Estimated coefficients (params):", fitted_model.params)
print("Coefficients as values:", fitted_model.params.values)
print("R-squared:", fitted_model.rsquared)

# Step 5: Visualize the data and fitted regression line
df['Data'] = 'Data'  # Adds a dummy 'Data' label to add to the plot's legend
fig = px.scatter(df, x='x', y='Y', color='Data', trendline='ols', title='Y vs. x')
# The `trendline='ols'` argument automatically fits and displays an OLS trendline

# Manually add trendline to the plot for comparison
fig.add_scatter(x=df['x'], y=fitted_model.fittedvalues,
                line=dict(color='blue'), name="trendline='ols'")

# Display plot (use `fig.show(renderer="png")` for GitHub or MarkUs submissions)
fig.show(renderer="png")
````

#### The Fitted Model Equation
The fitted Simple Linear Regression model is given by:

$
\hat{Y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i
$

where:

- $ \hat{\beta}_0 $ is the estimated intercept, derived from the data.
- $ \hat{\beta}_1 $ is the estimated slope coefficient, derived from the data.
- $ x_i $ represents each value in the predictor variable $ x $.

#### Using `fitted_model.params` for Predictions
- `fitted_model.params` contains the estimated values for $ \hat{\beta}_0 $ (intercept) and $ \hat{\beta}_1 $ (slope).
- For each observation $ x_i $ in the dataset, the predicted value $ \hat{Y}_i $ is computed by plugging $ x_i $ into the fitted equation above.

#### How `fitted_model.fittedvalues` is Derived
- Internally, `fitted_model.fittedvalues` applies the estimated parameters from `fitted_model.params` to each $ x_i $ in the data, resulting in an array of predicted values $ \hat{Y}_i $ for each observation.

### 6. Explain why the first expression below can be interpreted as "the proportion of variation in (outcome) Y explained by the model (i.e. _fitted_model.fittedvalues_)"; and therefore, why _fitted_model.rsquared_ can be interpreted as a measure of the accuracy of the model; and, therefore what the two _np.corrcoef(...)[0,1]\*\*2_ expressions capture in the context of _Simple Linear Regression models_.

1. `1-((Y-fitted_model.fittedvalues)**2).sum()/((Y-Y.mean())**2).sum()`
2. `fitted_model.rsquared`
3. `np.corrcoef(Y,fitted_model.fittedvalues)[0,1]**2`

<details class="details-example"><summary style="color:blue"><u>Further Guidance</u></summary>

> _**R-squared** is the "the proportion of variation in (outcome) $Y$ explained by the model ($\hat y_i$)" and is defined as_
>
> _$R^2 = 1 - \frac{\sum_{i=1}^n(Y_i-\hat y)^2}{\sum_{i=1}^n(Y_i-\bar Y)^2}$_
>
> _The visuzation provided in the previous problem can be used to consider $(Y_i-\bar Y)^2$ as the squared distance of the $Y_i$ to their sample average $\bar Y$ as opposed to the squared **residuals** $(Y_i-\hat y)^2$ which is the squared distance of the $Y_i$ to their fitted (predicted) values $Y_i$._
>

>

#### 1. The Expression Breakdown

**Proportion of Variation Explained**
- **Numerator**: $\sum (Y - \text{fitted\_model.fittedvalues})^2$ represents the **sum of squared residuals**. This measures how much the predicted values ($\text{fitted\_model.fittedvalues}$) deviate from the actual values $Y$. It indicates the unexplained variation in the model.
  
- **Denominator**: $\sum (Y - Y.\text{mean})^2$ represents the **total sum of squares** (SST), which measures the total variation in the observed values $Y$ relative to the mean of $Y$ This value quantifies the total variability present in the data.

The entire expression 

$
1 - \frac{\sum (Y - \text{fitted\_model.fittedvalues})^2}{\sum (Y - Y.\text{mean})^2}
$

thus calculates the proportion of the total variation in \(Y\) that is explained by the model. The closer this value is to 1, the more variance the model explains, indicating a better fit.

#### 2. R-squared Interpretation
- **R-squared $R^2$** is defined as:

$
R^2 = 1 - \frac{\sum (Y - \hat{Y})^2}{\sum (Y - \bar{Y})^2}
$

where $\hat{Y}$ represents the fitted values. Thus, \(fitted\_model.rsquared\) is simply a computational representation of the above expression and encapsulates the proportion of the variance explained by the model.

- Because $R^2$ gives a normalized measure of how well the independent variables explain the variability in the dependent variable, it serves as a metric for the **accuracy** of the model: a higher $R^2$ means a more accurate model.

#### 3. Correlation Coefficient Expressions
- The expression 

$
\text{np.corrcoef}(Y, \text{fitted\_model.fittedvalues})[0, 1]^2
$

captures the square of the Pearson correlation coefficient between $Y$ and the fitted values. This squared correlation reflects how well the predicted values (fitted values) relate to the actual values. If the correlation is high, this indicates a strong linear relationship, meaning the model is accurately capturing the trends in the data.

- Similarly, 

$
\text{np.corrcoef}(Y, x)[0, 1]^2
$

computes the squared correlation coefficient between the outcome variable $Y$ and the independent variable $x$ This value indicates how well $x$ alone explains the variation in $Y$.

````python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import plotly.express as px

# Load and filter the Old Faithful dataset
old_faithful = sns.load_dataset('geyser')
long_wait_limit = 71
long_wait = old_faithful[old_faithful.waiting > long_wait_limit]

# Initialize a list to collect slope coefficients from bootstrap samples
bootstrap_slopes = []

# Perform bootstrapping
n_bootstrap_samples = 1000  
for _ in range(n_bootstrap_samples):
    bootstrap_sample = long_wait.sample(n=len(long_wait), replace=True)
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slopes.append(model.params['waiting'])

# Visualize the distribution of bootstrap slope coefficients
fig = px.histogram(bootstrap_slopes, nbins=30, title="Bootstrapped Sampling Distribution of Slope Coefficients")
fig.show(renderer="png")
````

````python
# Parameters for the null hypothesis simulation
beta_0 = 1.65
beta_1 = 0
sigma = 0.37

# Initialize a list to collect slope coefficients under the null hypothesis
null_hypothesis_slopes = []

for _ in range(n_bootstrap_samples):
    simulated_y = beta_0 + beta_1 * long_wait['waiting'] + np.random.normal(0, sigma, len(long_wait))
    simulated_data = pd.DataFrame({'waiting': long_wait['waiting'], 'duration': simulated_y})
    model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    null_hypothesis_slopes.append(model.params['waiting'])

# Visualize the sampling distribution of slope coefficients under the null hypothesis
fig = px.histogram(null_hypothesis_slopes, nbins=30, title="Sampling Distribution of Slope Coefficients under H₀")
fig.show(renderer="png")
````

````python
# Calculate 95% confidence interval for the bootstrapped slopes
lower_bound = np.percentile(bootstrap_slopes, 2.5)
upper_bound = np.percentile(bootstrap_slopes, 97.5)

# Report if 0 is within the confidence interval
contains_zero = lower_bound <= 0 <= upper_bound
print(f"95% Confidence Interval: ({lower_bound:.4f}, {upper_bound:.4f})")
print(f"Does the 95% CI contain zero? {'Yes' if contains_zero else 'No'}")

# Compare with the p-value from the original fitted model
original_model = smf.ols('duration ~ waiting', data=long_wait).fit()
original_p_value = original_model.pvalues['waiting']
empirical_p_value = np.mean(np.array(bootstrap_slopes) >= original_model.params['waiting'])

print(f"Original Model p-value: {original_p_value:.4f}")
print(f"Empirical p-value from bootstrap: {empirical_p_value:.4f}")
````

````python
import numpy as np
import pandas as pd
import seaborn as sns

# Load dataset and define observed mean difference
old_faithful = sns.load_dataset('geyser')
old_faithful['kind'] = old_faithful['waiting'].apply(lambda x: 'short' if x < 68 else 'long')
observed_diff = old_faithful.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]

# Permutation test
n_permutations = 1000
permutation_diffs = []
for _ in range(n_permutations):
    shuffled_df = old_faithful.assign(kind_shuffled=old_faithful['kind'].sample(frac=1).values)
    perm_diff = shuffled_df.groupby('kind_shuffled')['duration'].mean().iloc[::-1].diff().values[1]
    permutation_diffs.append(perm_diff)

# Calculate p-value
p_value = np.mean([abs(diff) >= abs(observed_diff) for diff in permutation_diffs])
print(f"Observed mean difference: {observed_diff}")
print(f"Permutation test p-value: {p_value}")
````

````python
bootstrapped_diffs = []
n_bootstrap_samples = 1000
for _ in range(n_bootstrap_samples):
    boot_df = old_faithful.groupby('kind').apply(lambda x: x.sample(n=len(x), replace=True)).reset_index(drop=True)
    boot_diff = boot_df.groupby('kind')['duration'].mean().iloc[::-1].diff().values[1]
    bootstrapped_diffs.append(boot_diff)

# Calculate 95% bootstrap confidence interval
conf_interval = np.quantile(bootstrapped_diffs, [0.025, 0.975])
print(f"95% Bootstrap confidence interval for the difference in means: {conf_interval}")
````

#### Part (a): Explanation of Sampling Approaches
- Permutation Test:

    - This test creates a null distribution by randomly shuffling group labels and recalculating the mean difference between groups. By shuffling labels, we break any inherent association between the group label and duration.
    - The resulting distribution of mean differences represents what we’d expect if there were no actual difference between "short" and "long" groups.
- Bootstrap Confidence Interval:

    - In this method, we generate new samples by resampling within each group (sampling with replacement) and calculating the mean difference for each resampled dataset.
    - This allows us to estimate the variability in the mean difference without assuming normality, providing a direct estimate of the sampling distribution of the difference in means.
#### Part (b): Comparison with Indicator Variable Model Approach


- Similarities:

    - Both Approaches Aim to Assess Group Differences: All methods focus on understanding the difference in eruption duration between "short" and "long" groups.
    - Non-parametric Nature: The permutation and bootstrap approaches are non-parametric, meaning they don’t require the assumption of normality, similar to how the indicator model makes no assumption about within-group distributions.
- Differences:

    - Approach to Inference:
        - The indicator model relies on a regression framework and calculates the significance of the "kind" variable using a t-test, assuming normally distributed residuals.
        - The permutation test assesses significance by generating a null distribution through random label shuffling, without making assumptions about the underlying distributions.
        - The bootstrap method constructs a confidence interval by resampling within groups, providing a direct empirical estimate of the difference in means.
    - Type of Output:
        - The indicator variable model gives a p-value for the significance of the group difference.
        - The permutation test provides an empirical p-value based on the null distribution of shuffled differences.
        - The bootstrap method offers a confidence interval, giving a range within which we expect the true mean difference to lie.

> - and the $\epsilon_i$ **errors** are **unbiased** relative to the **expected value** of **outcome** $E[Y_i|x_i]=\beta_0 + \beta_1x_i$ (which is equivalently stated by saying that the mean of the **error distribution** is $0$, or again equivalently, that the **expected value** of the **errors** $E[\epsilon_i] = 0$)
