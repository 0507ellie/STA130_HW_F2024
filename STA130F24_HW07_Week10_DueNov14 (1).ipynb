### 1. Explain succinctly in your own words (but working with a ChatBot if needed)...<br>

1. the difference between **Simple Linear Regression** and **Multiple Linear Regression**; and the benefit the latter provides over the former
    - Simple Linear Regression (SLR) models the relationship between one predictor (independent variable) and an outcome (dependent variable) as a straight line. 
    - Multiple Linear Regression (MLR) extends this by including multiple predictors, allowing the model to capture relationships between several variables and the outcome.
    - Benefit: MLR can account for the influence of multiple factors simultaneously, improving model accuracy and interpretability by isolating each predictor's impact on the outcome while controlling for others.
2. the difference between using a **continuous variable** and an **indicator variable** in **Simple Linear Regression**; and these two **linear forms**
    - Continuous Variable in SLR: Represents a variable with numerical values where the effect is captured linearly.
    - Indicator Variable in SLR: Represents a binary condition. 

3. the change that happens in the behavior of the model (i.e., the expected nature of the data it models) when a single **indicator variable** is introduced alongside a **continuous variable** to create a **Multiple Linear Regression**; and these two **linear forms** (i.e., the **Simple Linear Regression** versus the **Multiple Linear Regression**)
    - Adding an **indicator variable** alongside a continuous variable in MLR allows different groups to have distinct intercepts while sharing the same slope (change rate) for the continuous predictor:

  **SLR Form:** 
  $$
  \text{outcome} = \beta_0 + \beta_1 \times \text{predictor} + \epsilon
  $$

  **MLR Form with Indicator:** 
  $$
  \text{outcome} = \beta_0 + \beta_1 \times \text{predictor} + \beta_2 \times 1(\text{group}) + \epsilon
  $$

    - **Behavior Change:** The indicator variable shifts the intercept based on group membership. For instance, if gender is an indicator variable, this form accounts for different baseline outcomes between genders but applies the same slope for the predictor (e.g., age).


4. the effect of adding an **interaction** between a **continuous** and an **indicator variable** in **Multiple Linear Regression** models; and this **linear form**
    - Adding an **interaction term** (product of a continuous and indicator variable) allows both the slope and intercept to differ across groups:

  $$
  \text{outcome} = \beta_0 + \beta_1 \times \text{predictor} + \beta_2 \times 1(\text{group}) + \beta_3 \times (\text{predictor} \times 1(\text{group})) + \epsilon
  $$

    - **Interpretation:** The interaction term \( \beta_3 \times (\text{predictor} \times 1(\text{group})) \) modifies the slope for one group, allowing the effect of the continuous predictor on the outcome to differ by group, giving a richer model fit.



5. the behavior of a **Multiple Linear Regression** model (i.e., the expected nature of the data it models) based only on **indicator variables** derived from a **non-binary categorical variable**; this **linear form**; and the necessarily resulting **binary variable encodings** it utilizes
    - When a **categorical predictor with multiple levels** is used (e.g., education level: high school, college, graduate), we represent it with \( K-1 \) binary indicator variables (one for each level except a baseline):

  $$
  \text{outcome} = \beta_0 + \beta_1 \times 1(\text{college}) + \beta_2 \times 1(\text{graduate}) + \epsilon
  $$

  Here, “high school” is the baseline (implicitly covered by \( \beta_0 \)), and college and graduate levels have specific coefficients for comparison.

- **Interpretation:** Each coefficient (\( \beta_1 \), \( \beta_2 \)) represents the difference from the baseline (high school), simplifying comparisons while preventing redundancy.

#### Outcome and Predictor Variables

- **Outcome Variable**: Sales of sports equipment (the amount sold).
- **Predictor Variables**:
  - **TV Advertising Budget**: Amount spent on TV ads.
  - **Online Advertising Budget**: Amount spent on online ads.

#### Need for Interaction

The **interaction** between TV and online advertising might be meaningful if spending on one type of advertising impacts the effectiveness of the other. For example, spending more on online ads could make TV ads more effective or vice versa. This is worth exploring with an interaction term in the model.

---

#### Linear Forms with and without Interaction

1. **Without Interaction**:
   - This form assumes that the effect of TV and online ads on sales is **independent** of each other:
   $$
   \text{Sales} = \beta_0 + \beta_1 \times \text{TV\_Budget} + \beta_2 \times \text{Online\_Budget} + \epsilon
   $$

2. **With Interaction**:
   - This form includes an **interaction term** to capture how TV and online ad spending might influence each other’s effectiveness:
   $$
   \text{Sales} = \beta_0 + \beta_1 \times \text{TV\_Budget} + \beta_2 \times \text{Online\_Budget} + \beta_3 \times (\text{TV\_Budget} \times \text{Online\_Budget}) + \epsilon
   $$

---

#### How to Use These Formulas

- **Without Interaction**: Predicts sales by adding the independent effects of TV and online budgets. This assumes that increasing one budget has the same effect on sales, regardless of the other.
- **With Interaction**: Predicts sales by considering how the combination of TV and online budgets affects sales. For example, higher spending on both types of ads could lead to a greater combined effect than just adding their separate effects.

---

#### Updating Formulas for Binary (High/Low) Advertising Budgets

If advertising budgets are **categorized as “high” or “low”**:

1. **Without Interaction**:
   $$
   \text{Sales} = \beta_0 + \beta_1 \times 1(\text{TV\_Budget\_High}) + \beta_2 \times 1(\text{Online\_Budget\_High}) + \epsilon
   $$

   Here, \(1(\text{TV\_Budget\_High})\) and \(1(\text{Online\_Budget\_High})\) are indicator variables (1 if high, 0 if low).

2. **With Interaction**:
   $$
   \text{Sales} = \beta_0 + \beta_1 \times 1(\text{TV\_Budget\_High}) + \beta_2 \times 1(\text{Online\_Budget\_High}) + \beta_3 \times (1(\text{TV\_Budget\_High}) \times 1(\text{Online\_Budget\_High})) + \epsilon
   $$

   The interaction term \( \beta_3 \times (1(\text{TV\_Budget\_High}) \times 1(\text{Online\_Budget\_High})) \) accounts for the combined effect when both budgets are high.

> _How do we simultaneously interpret **hypothesis testing** results regarding **coefficient estimates** based on **p-values** and **R-squared** "the proportion of variation in (outcome) $y$ explained by the model ($\hat y_i$)"? How can both be meaningfully understood at the same time? Do they address different aspects of a model?_
>    
> _As introduced in the previous homework, **R-squared** is_
>
> _$$R^2 = 1 - \frac{\sum_{i=1}^n(y_i-\hat y)^2}{\sum_{i=1}^n(y_i-\bar y)^2}$$_
>    
> _which describes the **explanatory power** of a model; whereas, **p-values** allow us to characterize **evidence against** a **null hypothesis**, and **coefficients** in a **multiple linear regression** context allow us to interpret the relationship between the **outcome** and a **predictor variable** "with all other **predictor variables** 'held constant'". Are these concepts thus contradictory or conflictual in some manner?_

|p-value|Evidence|
|-|-|
|$$p > 0.1$$|No evidence against the null hypothesis|
|$$0.1 \ge p > 0.05$$|Weak evidence against the null hypothesis|
|$$0.05 \ge p > 0.01$$|Moderate evidence against the null hypothesis|
|$$0.01 \ge p > 0.001$$|Strong evidence against the null hypothesis|
|$$0.001 \ge p$$|Very strong evidence against the null hypothesis|
    
> _In `formula='HP ~ Q("Sp. Def") * C(Generation)'` the `Q` stands for "quote" and is needed to access column names when they have a "space" in their name, while the `C` indicates a **categorical** use of what is actually an **integer** valued column. Despite technically being **continuous** numbers, **integer** often simply indicate categories which should not necessarily be treated as an incremental **continuous predictor variable**. Remember, a model such as $\beta_0 + \beta_1 x$ means for each unit increase in $x$ the outcome increases "on average" by $\beta_1$; so, if $x$ takes on the values `1` through `6` as the `Generation` **predictor variable** here does, then this means the average value for "Generation 1" must be $\beta_0 + \beta_1$ while for "Generation 2" it must be $\beta_0 + 2\times \beta_1$ (and so on up to "Generation 6" which must be $\beta_0 + 6\times \beta_1$). This might be a very strange restriction to place on something that is really actually a **categorical predictor variable**. You can see in the given model fit below how this six-level **categorical predictor variable** is actually appropriately treated in the specification of the **linear form** using "Generation 1" for the "baseline" and **binary indicators** to model the "contrast" ("offsets") for the other "Generations"; and, how these are in turn used in the context of the **interaction** considered by the model specification._

> _Recall from the previous week's homework that the **R-squared** "the proportion of variation in (outcome) $y$ explained by the model ($\hat y_i$)" quantity (re-introduced in the previous problem) can be calculated as `np.corrcoef(y,fitted_model.fittedvalues)[0,1]**2` (as well as several other ways in the **simple linear regression** context). The **squared correlation** between the outcome $y$ and it's **fitted values** $\hat y$ is the most generally useful formulation of **R-squared** since this can be use in the **multiple linear regression** context._
> 
> _This question therefore thus addresses the question of model **generalizability** on the basis of "in sample" and "out of sample" **model performance** (measured by **R-squared**)._
> 
> - _The **squared correlation** between the **outcomes** $y$ and their **fitted values** $\hat y$ is an "in sample" **model performance** metric since the $\hat y$ "predictions" for the $y$ **outcomes** are based on using those already **observed outcomes** to fit the model to generate the $\hat y$._  
> 
> - _If we instead calculate **squared correlation** between **outcomes** $y$ that were not used to fit the model and their corresponding $\hat y$ **predictions** (which are indeed now actually **predictions** as opposed to **fitted values**), then we are now  calculating an "out of sample" **model performance** metric._
> 
> _When an "out of sample" metric performs more poorly than a comparitive "in sample" metric, then the **predictions** of the **fitted model** are not **generalizing** to data being the dataset the model is fit on. In this case we say the model is **overfit** (to the data its fit was based on). The purpose of using different **training** and **testing** datasets is to consider "in sample" versus "out of sample" **model performance** in order to try to confirm that the model is not **overfit** and that the **predictions** do indeed seem to **generalizable** beyond the dataset used for **model fitting**._

### How model4_linear_form Creates New Predictors in the Design Matrix

- **model4_linear_form** specifies interactions between multiple variables, like `Attack`, `Defense`, `Speed`, `Legendary`, `Sp. Atk`, and `Sp. Def`.
- Each interaction term (e.g., `Attack * Defense * Speed`) creates additional columns in the "design matrix," `model4_spec.exog`.
- This **design matrix** is used to fit the model to predict the outcome, HP, by expanding combinations of these variables into new predictors.

### Why Multicollinearity Affects Out-of-Sample Generalization

- **Multicollinearity**: When predictors in the design matrix are highly correlated, making it hard for the model to distinguish their individual effects.
- **High Condition Number**: Indicates severe multicollinearity, shown here by extremely high values in `model4_fit.summary().tables[-1]`. Even after centering and scaling, the condition number remains high (e.g., \(1.20 \times 10^{16}\)), suggesting multicollinearity is still an issue.
- **Effect on Generalization**: High multicollinearity causes the model to fit well in-sample (training data) but poorly out-of-sample (test data). This occurs because the model is capturing noise or redundant patterns that do not generalize.

### Summary
- **model4_linear_form** creates complex interactions, increasing predictors in the design matrix.
- **Multicollinearity** in this design matrix limits the model’s ability to generalize, resulting in poor out-of-sample R-squared.

````python
import statsmodels.formula.api as smf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Assume `pokeaman` and `pokeaman_test` dataframes already exist with appropriate data

# Lists to store in-sample and out-of-sample R-squared values
in_sample_r2 = []
out_sample_r2 = []

# Define the formula for the model
model_formula = 'HP ~ Attack + Defense + Speed'  # Example formula

# Run loop for multiple iterations
for i in range(50):  # Run for 50 iterations
    # Randomly sample training data (e.g., select a different subset each time)
    train_data = pokeaman.sample(frac=0.8)  # 80% for training
    test_data = pokeaman.drop(train_data.index)  # Remaining 20% for testing

    # Fit the model on the training data
    model = smf.ols(formula=model_formula, data=train_data)
    model_fit = model.fit()

    # Calculate in-sample R-squared
    in_sample_r2.append(model_fit.rsquared)

    # Predict on test data for out-of-sample R-squared
    y = test_data['HP']
    yhat = model_fit.predict(test_data)
    out_sample_r2.append(np.corrcoef(y, yhat)[0, 1] ** 2)

# Convert results to a DataFrame for easy handling and plotting
results = pd.DataFrame({
    'In-sample R-squared': in_sample_r2,
    'Out-of-sample R-squared': out_sample_r2
})

# Plot the results to visualize distribution of R-squared values
plt.figure(figsize=(10, 6))
plt.plot(results['In-sample R-squared'], label="In-sample R-squared", marker='o')
plt.plot(results['Out-of-sample R-squared'], label="Out-of-sample R-squared", marker='o')
plt.xlabel('Iteration')
plt.ylabel('R-squared')
plt.title('In-sample vs Out-of-sample R-squared over Multiple Iterations')
plt.legend()
plt.show()
````
